### Scatter Plots
- great for showing the relationship between two variables as we can directly see the raw distribution of the data

```python
# Generating Random Data
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
```

- **alpha = 0.2**: The optional alpha input parameter (optional) of the scatter method allows us to adjust the transparency level so that we can view overlapping data points.
- **cmap='viridis'**: Matplotlib provides a wide range of colormaps, an optional parameter, for creating visually appealing plots. For example, you can set cmap=plt.cm.Blues, cmap=plt.cm.autumn, or cmap=plt.cm.gist_earth
- The color argument, **c**, is automatically mapped to a color scale, and the size argument, **s**, is in pixels

- If this was a real data set, the x, y location of each point can be used to represent two distinct features. The size of the points can be used to represent a third feature, and the color mapping can be used to represent different classes or groups of the data points

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis') 
plt.colorbar();  # To show the color scale
```

### Line Plots

- Use the plt.plot() function to plot population against time.

```python
# 2. Getting the Data
year = [1970, 1980, 1990, 2000, 2010, 2020]
pop_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
pop_B = [44.4, 55.6, 69.7, 87.1, 95.4, 100.5]

# 3. Visualising the Data
plt.plot(year, pop_A, color='g')
plt.plot(year, pop_B, color='r')
plt.xlabel('Countries')
plt.ylabel('Population in million')
plt.title('Populations over the years')
```

### Histograms
- histogram is a plot where data is split into bins. A bin is a range of values or intervals. For each bin, we get the count of how many values fall into it. The x-axis represents bin ranges while y-axis shows frequency.
- We can plot histograms using the **hist()** function

```python
data = np.random.randn(1000)
plt.hist(x=data);
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5,
                histtype='stepfilled', color='steelblue',
                edgecolor='none');
```

### Bar Plots

```python
# Input data
labels = ('Python', 'C++', 'Java', 'Perl', 'C#')
num_users = [12,8,15,4,6]
index = np.arange(len(labels))
index

# Plotting
plt.bar(index, num_users, align='center', alpha=0.5)
plt.xticks(index, labels)
plt.xlabel('Language')
plt.ylabel('Num of Users')
plt.title('Programming language usage')
```
